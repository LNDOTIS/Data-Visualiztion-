```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import colorcet as cc
%reload_ext autoreload
%autoreload 2
```

```python
gini_df = pd.read_csv('swiid9_8_summary.csv')

# Filter Finland data and select relevant columns
finland_df = gini_df[gini_df['country'] == 'Finland'][['year', 'gini_disp']]

sns.set_style("white")
# Plot the data
plt.figure(figsize=(10, 5))
sns.lineplot(data=finland_df, x='year', y='gini_disp', marker='o', color='r')

# Customize the plot
plt.title('Income Inequality in Finland (Gini Coefficient Over Time)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Gini Coefficient (gini_disp)', fontsize=12)
plt.show()
```
